# 6. **Outlier Detection:**

```python
numerical_columns = ['Age', 'Education', 'Usage', 'Fitness', 'Income', 'Miles']

for column in numerical_columns:
    Q1 = df[column].quantile(0.25)  # First quartile (25th percentile)
    Q3 = df[column].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1                  # Interquartile Range
    lower_bound = Q1 - 1.5 * IQR    # Lower bound
    upper_bound = Q3 + 1.5 * IQR    # Upper bound

    # Identifying outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    print(f"Column: {column}")
    print(f"  Lower Bound: {lower_bound:.2f}, Upper Bound: {upper_bound:.2f}")
    print(f"  Number of Outliers: {len(outliers)}")
    print("-" * 40)
```

What percent of customers have purchased KP281, KP481, or KP781?

# **Product â€“ Gender**
* Percentage of a Male customer purchasing a treadmill

```python
male_percentage = (df['Gender'].value_counts(normalize=True)['Male'] * 100).round(2)
print(f"Percentage of Male customers purchasing a treadmill: {male_percentage}%")
```

* #Percentage of a Female customer purchasing KP781 treadmill

```python
female_kp781_percentage = ((df['Gender'] == 'Female') & (df['Product'] == 'KP781')).sum() / len(df) * 100
female_kp781_percentage = round(female_kp781_percentage, 2)
print(f"Percentage of Female customers purchasing KP781 treadmill: {female_kp781_percentage}%")
```

* Probability of a customer being a Female given that Product is
KP281

```python
kp281_female_prob = (df[(df['Product'] == 'KP281') & (df['Gender'] == 'Female')].shape[0]) / (df['Product'] == 'KP281').sum()
kp281_female_prob = round(kp281_female_prob * 100, 2)
print(f"Probability of a customer being Female given that Product is KP281: {kp281_female_prob}%")
```
